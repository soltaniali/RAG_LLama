```python
def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = []

    for para in doc.paragraphs:
        full_text.append(para.text)

    for table in doc.tables:
        text=""
        for row in table.rows:
            for cell in row.cells:
                text +=(cell.text) + " "
        full_text.append(text)

    return full_text
```

```python
def extract_questions_from_docx(file_path, column_name):
    # Load the docx file
    doc = docx.Document(file_path)

    # Initialize a list to hold the questions
    questions = []

    # Iterate through the tables in the document
    for table in doc.tables:
        # Check the first row for the column names
        headers = [cell.text for cell in table.rows[0].cells]
        if column_name in headers:
            # Get the index of the "سوال" column
            column_index = headers.index(column_name)

            # Iterate through the rows and extract the data from the "سوال" column
            for row in table.rows[1:]:  # Skip the header row
                cell = row.cells[column_index]
                questions.append(cell.text)

    # Convert the list of questions to a Pandas DataFrame
    df = pd.DataFrame(questions, columns=[column_name])

    return df
```

```python
# Load question documents
questions1_df = extract_questions_from_docx("questions 1.docx","سوال")
questions2_df = pd.read_excel("questions 2.xlsx").iloc[1:]
```

```python
from transformers import AutoTokenizer
model_name="NousResearch/Meta-Llama-3-8B-Instruct"
tokenizer = AutoTokenizer.from_pretrained(
    # "meta-llama/Meta-Llama-3-8B-Instruct"
    model_name
)

stopping_ids = [
    tokenizer.eos_token_id,
    tokenizer.convert_tokens_to_ids("<|eot_id|>"),
]
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
from transformers import BitsAndBytesConfig
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
)

llm = HuggingFaceLLM(
    model_name=model_name,
    model_kwargs={"quantization_config": quantization_config},
    generate_kwargs={
        "do_sample": True,
        "temperature": 0.6,
        "top_p": 0.9,
    },
    tokenizer_name=model_name,
    stopping_ids=stopping_ids,
)
# llm = AutoModelForCausalLM.from_pretrained(model_name , quantization_config=quantization_config,
#     device_map="auto")
```
